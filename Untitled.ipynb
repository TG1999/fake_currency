```python
import math
```

```python
import numpy as np
```

```python
import cv2
```

```python
image = cv2.imread('1.jpg')
```

```python
cv2.imshow("image", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
gray
```

```python
cv2.imshow("image", gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
bi = cv2.bilateralFilter(gray, 5, 75, 75)
```

```python
bi
```

```python
dst = cv2.cornerHarris(bi, 2, 3, 0.04)
```

```python
dst
```

```python
mask = np.zeros_like(gray)
```

```python
mask
```

```python
mask[dst>0.01*dst.max()] = 255
```

```python
image[dst > 0.01 * dst.max()] = [0, 0, 255]
```

```python
cv2.imshow("image", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
coor = np.argwhere(mask)
```

```python
coor_list = [l.tolist() for l in list(coor)]
```

```python
coor_tuples = [tuple(l) for l in coor_list]
```

```python
thresh = 50

def distance(pt1, pt2):
    (x1, y1), (x2, y2) = pt1, pt2
    dist = math.sqrt( (x2 - x1)**2 + (y2 - y1)**2 )
    return dist

coor_tuples_copy = coor_tuples

i = 1    
for pt1 in coor_tuples:

    print(' I :', i)
    for pt2 in coor_tuples[i::1]:
        print(pt1, pt2)
        print('Distance :', distance(pt1, pt2))
        if(distance(pt1, pt2) < thresh):
            coor_tuples_copy.remove(pt2)      
    i+=1
```

```python
img2 = image.copy()
for pt in coor_tuples:
    cv2.circle(img2, tuple(reversed(pt)), 3, (0, 0, 255), -1)
```

```python
cv2.imshow("image", img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
